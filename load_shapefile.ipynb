```python
import pandas as pd
from shapely.geometry import Point

business_licenses = pd.read_excel("Atlanta_Business_License_Records_2025.xlsx")
food_services = business_licenses[business_licenses['license_classification'] == 'Accommodation and Food Services']
food_services = food_services.dropna(subset=['latitude', 'longitude'])
```

```python
import matplotlib.pyplot as plt

valid_food_services = food_services[
    (food_services['latitude'] >= 33.5) & 
    (food_services['latitude'] <= 34.0) &
    (food_services['longitude'] >= -84.7) & 
    (food_services['longitude'] <= -84.0)
].copy()

geometry = [Point(xy) for xy in zip(valid_food_services['longitude'], valid_food_services['latitude'])]
food_services_geo = gpd.GeoDataFrame(valid_food_services, geometry=geometry, crs='EPSG:4326')
food_services_geo = food_services_geo.to_crs(atlanta_city.crs)

fig, ax = plt.subplots(figsize=(14, 12))
atlanta_city.plot(ax=ax, color="lightgrey", edgecolor="black", label="Atlanta City Limits")
food_services_geo.plot(ax=ax, marker="o", color="red", markersize=40, alpha=0.7, label="Licensed Food Services")
ax.set_title('Licensed Food Service Businesses in Atlanta', fontsize=14, fontweight='bold')
ax.legend(loc='upper right', fontsize=10)
plt.tight_layout()
```
